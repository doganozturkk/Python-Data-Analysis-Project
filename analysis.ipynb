# Movie Data Analysis
This notebook demonstrates analysis of a movie dataset.

```
import pandas as pd
import matplotlib.pyplot as plt

movies = pd.DataFrame({
    'Title': ['Movie A','Movie B','Movie C','Movie D','Movie E'],
    'Year': [2010,2012,2015,2018,2020],
    'Rating': [7.2,6.5,8.1,7.8,6.9],
    'Votes': [12000,8000,25000,15000,9000]
})
movies
```

```
plt.figure(figsize=(6,4))
plt.plot(movies['Year'], movies['Rating'])
plt.xlabel('Year')
plt.ylabel('Rating')
plt.title('Ratings Over Years')
plt.show()
```
